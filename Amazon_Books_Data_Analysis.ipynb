```python
#Visulaizing According to the genres:
fig_1 = plt.figure(figsize=(10,10),dpi=100)
axes_1 = fig_1.add_axes([0.2,0.2,0.3,0.3])
labels=df["Genre"].unique()
values=[x for x in df["Genre"].value_counts()]
bars=axes_1.bar(labels,values)
```

```python
plt.xlabel("User Ratings")
plt.title("Ratings of the Books")
plt.hist(df["User Rating"], bins=20, density=True, stacked=True)
```

```python
#Rating vs the Price of the Books:
plt.xlabel("User Ratings")
plt.ylabel("Book Price")
plt.title("Ratings vs Book Price")
plt.scatter(df["User Rating"], df["Price"], c ="black")
```

```python
df=df.sort_values(by="User Rating",ascending=False)
df.head(10)
```

```python
#Whick Books are Available for free?
df[df["Price"] == 0]
```

```python
#Which books have a particular rating?
df[df["User Rating"] == 4.2]
```
